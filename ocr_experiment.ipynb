```python
import layoutparser as lp
model = lp.Detectron2LayoutModel('lp://PubLayNet/faster_rcnn_R_50_FPN_3x/config',
                                 extra_config=["MODEL.ROI_HEADS.SCORE_THRESH_TEST", 0.8],
                                 label_map={0: "Text", 1: "Title", 2: "List", 3:"Table", 4:"Figure"})
```

```python
# First, ensure the correct library (PyMuPDF) is installed and the incorrect 'fitz' is removed.

import fitz
# Ensure cv2 is imported before use
import cv2

file_path = "./form_scanned.pdf"
dpi = 1000  # choose desired dpi here
zoom = dpi / 72  # zoom factor, standard: 72 dpi
magnify = fitz.Matrix(zoom, zoom)  # magnifies in x, resp. y direction
doc = fitz.open(file_path)  # open document
for page in doc:
    pix = page.get_pixmap(matrix=magnify)  # render page to an image
    pix.save(f"page-{page.number}.png")

img = cv2.imread("./page-0.png", cv2.IMREAD_GRAYSCALE)


    # Increase contrast and apply adaptive thresholding (Binarization)
    # This helps separate text from the form lines
processed_img = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                          cv2.THRESH_BINARY, 11, 2)

cv2.imwrite('cleaned_form.png', processed_img)

```

```python
print(processed_img.shape)
# Convert grayscale processed_img to a 3-channel image for detectron2 and drawing
processed_img_bgr = cv2.cvtColor(processed_img, cv2.COLOR_GRAY2BGR)

layout = model.detect(processed_img_bgr)
    # Detect the layout of the input image

lp.draw_box(processed_img_bgr, layout, box_width=30)
```

```python
ocr_agent = lp.TesseractAgent(languages='eng')
```

```python
text_blocks = lp.Layout([b for b in layout if b.type=='Text'])
figure_blocks = lp.Layout([b for b in layout if b.type=='Figure'])




for block in text_blocks:
    segment_image = (block
                       .pad(left=5, right=5, top=5, bottom=5)
                       .crop_image(img))
        # add padding in each image segment can help
        # improve robustness

    text = ocr_agent.detect(segment_image)
    block.set(text=text, inplace=True)

for txt in text_blocks.get_texts():
    print(txt, end='\n---\n')

```
